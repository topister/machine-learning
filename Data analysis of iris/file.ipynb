```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
```

```python
list(iris_data.target_names)
```

```python
import pandas as pd
```

```python
# Create DataFrame for features (explanatory variables)
X = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
```

```python
# Create DataFrame for target (objective variable)
y = pd.DataFrame(data=iris_data.target, columns=['Species'])
```

```python
# Display first few rows of X and y to verify
print("Explanatory Variables (X):")
X.head()
```

```python
# Rename the columns of X
X.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
# Combine X and y into a single DataFrame df
df = pd.concat([X, y], axis=1)
```

### Number of Instances: 

The dataset contains a total of 150 instances, with 50 instances for each species of iris.

### Features: 

There are four features measured for each iris specimen:

Sepal length (in cm)

Sepal width (in cm)

Petal length (in cm)

Petal width (in cm)

### Target Variable (Labels): 

The target variable is the species of iris, which is categorized into three classes:

Iris setosa

Iris versicolor

Iris virginica

#### Explanation of .loc[] and .iloc[]:
.loc[] is primarily label-based indexing. It selects data based on row and column labels.

.iloc[] is primarily integer-based indexing. It selects data based on integer location of rows and columns.


### Creating a diagram


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# 1. Pie chart of the number of samples per label (with percentages)
plt.figure(figsize=(8, 6))
labels = df['Species'].value_counts().index
sizes = df['Species'].value_counts().values
explode = (0.1, 0, 0)  # explode the 1st slice
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Number of Samples per Label')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

Box plots provide a concise summary of the distribution, including outliers, quartiles, and median, in a compact visual form. They are useful for comparing distributions across different categories.

Violin plots offer a more detailed view of the distribution by also displaying the kernel density estimation plot, which shows the probability density at different values. They provide a more nuanced understanding of the data distribution, especially when dealing with multimodal or skewed distributions.

### Confirming the relationship between features

```python
# 1. Scatter plots of all combinations of features, color-coded by type
sns.pairplot(df, hue='Species', diag_kind='hist')
plt.suptitle('Scatter Plots of All Feature Combinations')
plt.show()
```

```python
# 2. Scatterplot matrix displaying all combinations of features at once
sns.pairplot(df)
plt.suptitle('Scatterplot Matrix of All Features')
plt.show()
```

```python
# 3. Correlation coefficient matrix for 4 features
correlation_matrix = df.corr()
correlation_matrix
```

```python
# 4. Heatmap of the correlation coefficient matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Coefficient Matrix of Features')
plt.show()
```

### Explaining the results of visualized graphs and tables


### Scatter Plots:

In the scatter plots, each point represents an iris sample, with the x and y axes representing different features (sepal length, sepal width, petal length, petal width). The points are color-coded by species (Iris setosa, Iris versicolor, Iris virginica).
From the scatter plots, we can observe distinct clusters for each species, indicating that different species have different feature distributions. For example, Iris setosa typically has smaller sepal and petal dimensions compared to Iris versicolor and Iris virginica.
We can also notice that some features have strong linear relationships with each other, such as petal length and petal width, indicating a high degree of correlation between these features.

### Correlation Coefficient Matrix:

The correlation coefficient matrix quantifies the strength and direction of linear relationships between pairs of features and the target variable (species).
Values close to 1 or -1 indicate strong positive or negative correlations, respectively, while values close to 0 indicate weak or no correlation.
From the correlation matrix, we can observe that petal length and petal width have the highest positive correlation with the species (0.949 and 0.957, respectively), followed by sepal length (0.783). This suggests that these features are highly informative for distinguishing between iris species.

### Heatmap of Correlation Coefficient Matrix:

The heatmap visualizes the correlation coefficient matrix, with brighter colors indicating stronger correlations.
We can observe strong positive correlations between petal length and petal width (0.96), sepal length and petal length (0.87), as well as sepal length and petal width (0.82).
These strong correlations indicate that these features tend to increase or decrease together, which is consistent with what we observed in the scatter plots.
