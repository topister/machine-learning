```python
import numpy as np

def MSE(y_pred, y):
    """
    Calculation of mean square error

    Parameters
    ----------
    y_pred : of the following form. ndarray, shape (n_samples,)
      Estimated value
    y : of the following form. ndarray, shape (n_samples,)
      Correct answer value

    Returns
    ----------
    mse : numpy.float
      Mean squared error
    """
    # Compute mean squared error
    mse = np.mean((y_pred - y) ** 2)
    return mse
```

#### 【problem 5】Objective function

```python
import numpy as np

class ScratchLinearRegression():
    """
    Scratch implementation of linear regression
    
    Parameters
    ----------
    num_iter : int
      Number of iterations
    lr : float
      Learning rate
    no_bias : bool
      True if no bias term is included
    verbose : bool
      True to output the learning process
    
    Attributes
    ----------
    self.coef_ : of the following form. ndarray, shape (n_features,)
      Parameters
    self.loss : of the following form. ndarray, shape (self.iter,)
      Record losses on training data
    self.val_loss : of the following form. ndarray, shape (self.iter,)
      Record loss on validation data
    """
    
    def __init__(self, num_iter, lr, no_bias, verbose):
        # Record hyperparameters as attributes
        self.iter = num_iter
        self.lr = lr
        self.no_bias = no_bias
        self.verbose = verbose
        # Prepare an array to record the loss
        self.loss = np.zeros(self.iter)
        self.val_loss = np.zeros(self.iter)
        
    def fit(self, X, y, X_val=None, y_val=None):
        """
        Learn linear regression. If validation data is entered, the loss and accuracy for it are also calculated for each iteration.
        Parameters
        ----------
        X : of the following form. ndarray, shape (n_samples, n_features)
            Features of training data
        y : of the following form. ndarray, shape (n_samples, )
            Correct answer value of training data
        X_val : of the following form. ndarray, shape (n_samples, n_features)
            Features of verification data
        y_val : of the following form. ndarray, shape (n_samples, )
            Correct value of verification data
        """
        if not self.no_bias:
            X = np.hstack((np.ones((X.shape[0], 1)), X))  # Add bias term to X
            if X_val is not None:
                X_val = np.hstack((np.ones((X_val.shape[0], 1)), X_val))  # Add bias term to validation data
        
        # Initialize parameters
        self.coef_ = np.random.randn(X.shape[1])

        for i in range(self.iter):
          # Compute predictions
          y_pred = self._linear_hypothesis(X)
          # Compute error
          error = y_pred - y
          # Update parameters using gradient descent
          self._gradient_descent(X, error)
          # Compute loss
          self.loss[i] = np.mean((error ** 2) / 2)
    
          if self.verbose:
              print(f"Iteration {i+1}/{self.iter}, Loss: {self.loss[i]}")
              print("Model parameters:", self.coef_)
        
        
    
    def predict(self, X):
      """
      Estimate using linear regression.
      Parameters
      ----------
      X : of the following form. ndarray, shape (n_samples, n_features)
          sample
      Returns
      -------
          of the following form. ndarray, shape (n_samples, 1)
          Estimated result by linear regression
      """
      if not self.no_bias:
          X = np.hstack((np.ones((X.shape[0], 1)), X))  # Add bias term to X
      y_pred = self._linear_hypothesis(X)
      # Handle NaN values
      if np.isnan(y_pred).any():
          raise ValueError("Model produced NaN values in predictions.")
      return y_pred

    
    def _linear_hypothesis(self, X):
        """
        Compute a linear hypothetical function

        Parameters
        ----------
        X : of the following form. ndarray, shape (n_samples, n_features)
          Training data

        Returns
        -------
          of the following form. ndarray, shape (n_samples, 1)
          Estimated result by linear hypothetical function

        """
        return np.dot(X, self.coef_)
    
    def _gradient_descent(self, X, error):
      """
      Update parameters using gradient descent

      Parameters
      ----------
      X : of the following form. ndarray, shape (n_samples, n_features)
        Training data
      error : of the following form. ndarray, shape (n_samples, 1)
        Difference between predicted and actual values

      """

      # Update parameters
      new_coef = self.coef_ - self.lr * np.mean(X * error[:, np.newaxis], axis=0)
      # Separate update for the bias term
      new_coef[0] = self.coef_[0] - self.lr * np.mean(error)
      # Check if parameters contain NaN values
      if np.isnan(new_coef).any():
          raise ValueError("Model parameters contain NaN values.")
      # Update parameters only if they are finite
      if np.isfinite(new_coef).all():
          self.coef_ = new_coef

```

```python
import numpy as np

# Define a function for feature scaling using standardization
def standardize_features(X):
    """
    Standardize features to have zero mean and unit variance.

    Parameters
    ----------
    X : ndarray
        Input feature matrix with shape (n_samples, n_features).

    Returns
    -------
    X_scaled : ndarray
        Standardized feature matrix.
    """
    # Calculate mean and standard deviation of each feature
    mean = np.mean(X, axis=0)
    std_dev = np.std(X, axis=0)
    
    # Avoid division by zero for features with zero standard deviation
    std_dev[std_dev == 0] = 1
    
    # Standardize features
    X_scaled = (X - mean) / std_dev
    
    return X_scaled

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some synthetic data
np.random.seed(0)
m = 100  # Number of examples
X = 2 * np.random.rand(m, 1)
y = 4 + 3 * X + np.random.randn(m, 1)

# Add the bias term (intercept) to each instance
X_b = np.c_[np.ones((m, 1)), X]

# Compute the cost function for different values of theta
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1 / (2 * m)) * np.sum(np.square(predictions - y))
    return cost

# Values of theta0 and theta1 to try
theta0_vals = np.linspace(0, 8, 100)
theta1_vals = np.linspace(0, 5, 100)
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

# Fill the cost function values
for i in range(len(theta0_vals)):
    for j in range(len(theta1_vals)):
        theta = np.array([theta0_vals[i], theta1_vals[j]])
        J_vals[i, j] = compute_cost(X_b, y, theta)

# Plot the cost function
theta0_vals, theta1_vals = np.meshgrid(theta0_vals, theta1_vals)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(theta0_vals, theta1_vals, J_vals.T, cmap='viridis')

ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'Cost $J(\theta)$')
ax.set_title('Cost Function Surface')

plt.show()
```

Because the cost function 𝐽(𝜃) in linear regression is convex, it does not have local minima, only a single global minimum. Therefore, gradient descent will always converge to this optimal solution, provided a suitable learning rate 𝛼 and sufficient iterations. This is why the problem of local optima does not affect linear regression.
