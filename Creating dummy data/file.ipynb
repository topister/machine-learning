## Creating random numbers


```python
import numpy as np

np.random.seed(0)
mean = (-3, 0)
covariance_matrix = [[1.0, 0.8], [0.8, 1.0]]
data_1 = np.random.multivariate_normal(mean, covariance_matrix, size=500)
data_1
```

## Visualization by scatter plot


```python
import matplotlib.pyplot as plt

plt.scatter(data_1[:, 0], data_1[:, 1])
plt.title('Scatter Plot of Data from Problem 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

## Visualization by histogram

```python
# Plot histogram for dimension 1
plt.hist(data_1[:, 0], bins=30, alpha=0.5, label='Dimension 1')

# Plot histogram for dimension 2
plt.hist(data_1[:, 1], bins=30, alpha=0.5, label='Dimension 2')

# Set title
plt.title('Histogram of Data from Problem 1')

# Set labels for x and y axes
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

## Addition of data and visualization by scatter plot

```python
# Define mean and covariance matrix for Problem 4 data
mean_2 = (0, -3)
covariance_matrix = [[1.0, 0.8], [0.8, 1.0]]

# Generate random data for Problem 4
data_2 = np.random.multivariate_normal(mean_2, covariance_matrix, size=500)

# Plot scatter plot for Problem 1 data
plt.scatter(data_1[:, 0], data_1[:, 1], label='Problem 1 (0)')

# Plot scatter plot for Problem 4 data
plt.scatter(data_2[:, 0], data_2[:, 1], label='Problem 4 (1)')

# Set title
plt.title('Scatter Plot of Data from Problem 1 and 4')

# Set labels for x and y axes
plt.xlabel('X')
plt.ylabel('Y')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

## Data combination

```python
combined_data = np.concatenate((data_1, data_2), axis=0)
combined_data
```

## Labeling

```python
labels = np.zeros((1000, 1))
labels[500:] = 1
labeled_data = np.hstack((combined_data, labels))
labeled_data
```
