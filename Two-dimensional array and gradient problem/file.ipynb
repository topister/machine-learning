```python
import numpy as np
import matplotlib.pyplot as plt

# Problem 1: Linear function
x = np.arange(-50, 50.1, 0.1)
y = 0.5 * x + 1

# Problem 2: Array combination
array_xy = np.column_stack((x, y))

# Problem 3: Find the gradient
gradient = np.diff(y) / np.diff(x)

# Problem 4: Draw a graph
plt.figure(figsize=(10, 5))

# Linear function plot
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.legend()

# Gradient plot
plt.subplot(1, 2, 2)
plt.plot(x[:-1], gradient, label='Gradient')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.title('Gradient')
plt.legend()

plt.tight_layout()
plt.show()

# Problem 5: Python functionalization
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)
    array_xy = np.column_stack((x, y))
    gradient = np.diff(y) / np.diff(x)
    return array_xy, gradient

def function1(array_x):
    return 0.5 * array_x + 1

array_xy, gradient = compute_gradient(function1)
print("Gradient:", gradient)

# Problem 6: Find the minimum value
min_y = np.min(array_xy[:, 1])
min_index = np.argmin(array_xy[:, 1])
min_slope_before = gradient[min_index - 1]
min_slope_after = gradient[min_index]

print("Minimum y:", min_y)
print("Slope before minimum y:", min_slope_before)
print("Slope after minimum y:", min_slope_after)
```
