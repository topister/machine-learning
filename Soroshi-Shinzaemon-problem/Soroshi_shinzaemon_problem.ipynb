```python
import matplotlib.pyplot as plt
```

```python

def compute_sorori_shinzaemon(n_days=100):
    """A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days: int
        Number of days to get rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice you get one day (list length equals n_days)
    list_total_grains : list
        List of total number of rice received by a certain day (list length is equal to n_days)
    """

    list_n_grains = []    # List to store the number of rice grains received each day
    list_total_grains = []    # List to store the total number of rice grains received by each day

    total_grains = 0    # Variable to store the total number of rice grains

    for day in range(1, n_days + 1):
        grains_today = 2**(day - 1)    # Calculate the number of grains received on the current day
        total_grains += grains_today    # Update the total number of grains
        list_n_grains.append(grains_today)    # Add the number of grains received today to the list
        list_total_grains.append(total_grains)    # Add the total number of grains to the list

    # Plotting the line graph
    plt.plot(range(1, n_days + 1), list_n_grains, label='Daily Grains')
    plt.plot(range(1, n_days + 1), list_total_grains, label='Total Grains')
    plt.xlabel('Days')
    plt.ylabel('Number of Rice Grains')
    plt.title('Sorori Shinzaemon Problem')
    plt.legend()
    plt.show()

    return list_n_grains, list_total_grains
```

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=100)

```python
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=100)
```

```python
def calculate_days_to_live(num_grains, num_people):
    """Calculate how many days people can live with the given amount of rice.

    Parameters
    ----------
    num_grains: int
        Number of rice grains
    num_people: int
        Number of people

    Returns
    -------
    days_to_live: int
        Number of days people can live with the given amount of rice
    """

    # This is an assumption of an average person needs around 400 grams of rice per day
    grains_per_day_per_person = 400
    days_to_live = num_grains / (num_people * grains_per_day_per_person)

    return int(days_to_live)
```

```python

num_grains_example = list_total_grains[-1]  # Using the total grains on the 100th day as an example
num_people_example = 5  # Example number of people
days_to_live_example = calculate_days_to_live(num_grains_example, num_people_example)

print(f"With {num_grains_example} grains, {num_people_example} people can live for approximately {days_to_live_example} days.")
```
